```python
import numpy as np
import math
import os
import cv2
import matplotlib.pyplot as plt
```

```python
class BouncingBall:
  def __init__(self, dt, w, l, r, cx, cy, vx, vy):
    self.dt = dt
    self.r = r
    self.w = w
    self.l = l
    self.cx = cx
    self.cy = cy
    self.vx = vx
    self.vy = vy

  def update_c(self):
    next_cx = self.cx+self.vx*dt
    if l < next_cx:
        next_cx = l-(next_cx-l)
    if 0 > next_cx:
        next_cx = -next_cx
    self.cx = next_cx
    
    next_cy = self.cy+self.vy*dt
    if w < next_cy:
        next_cy = w-(next_cy-w)
        self.vy = -self.vy
        
    if 0 > next_cy:
        next_cy = -next_cy
        self.cy = next_cy


def ring(r):
    cir = np.zeros((2*r+1, 2*r+1))
    idx = np.array(range(-r, r+1, 1))
    for x in idx:
        for y in idx:
            d = math.sqrt(x*x+y*y)
            if d <= r and d >= r-1:
                cir[r+x, r+y] = 1    

def circle(r):
    cir = np.zeros((2*r+1, 2*r+1))
    idx = np.array(range(-r, r+1, 1))
    for x in idx:
        for y in idx:
            d = math.sqrt(x*x+y*y)
            if d <= r:
                cir[r+x, r+y] = 1         
                
    return cir

def update_pixel_frame(bball):
    mat = np.zeros((bball.w,bball.l))
    cir = circle(bball.r)
    mat[bball.cy-bball.r:bball.cy+bball.r+1, bball.cx-bball.r:bball.cx+bball.r+1] = cir
    
    
    print(mat)
    
    return mat*255

    
    
```
